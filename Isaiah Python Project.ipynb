```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg
```

Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.set(style="whitegrid")
sns.relplot(data=mpg, x='horsepower', y='mpg', kind='scatter', height=6, aspect=1.5)
plt.title('Relationship between Horsepower and Fuel Economy (MPG)')
plt.show()
```

Create a table summarizing fuel economy by country of origin

```python
mpg_summary = mpg.groupby('origin')['mpg'].mean().reset_index()
mpg_summary.columns = ['Country of Origin', 'Average MPG']
```

```python
mpg_summary
```

Create a visual showing how fuel economy has changed over time

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=mpg, x='model_year', y='mpg', ci=None)
plt.title('Change in Fuel Economy (MPG) Over Time')
plt.xlabel('Model Year')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

Create two other visuals or tables related to fuel economy.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=mpg, x='origin', y='mpg')
plt.title('Distribution of Fuel Economy by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=mpg, x='cylinders', y='mpg', ci=None)
plt.title('Average Fuel Economy by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Miles per Gallon (MPG)')
plt.show()
```

Box plot of fuel economy by country of origin:
Cars from Japan and Europe generally have higher fuel economy (MPG) compared to those from the USA.
The USA has a wider range of fuel economy, with many vehicles showing lower MPG, reflecting the larger number of high-powered cars in this region.

Bar plot of fuel economy by number of cylinders:
Vehicles with 4 cylinders have the highest average fuel economy, which aligns with the fact that smaller engines are generally more fuel-efficient.
As the number of cylinders increases, the average MPG drops significantly, with 8-cylinder vehicles having the lowest fuel economy.

These insights suggest that fuel economy is influenced by both the engine size (number of cylinders) and the origin of the vehicles, with smaller engines and foreign cars generally being more fuel-efficient.
